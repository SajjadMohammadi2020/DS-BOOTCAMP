```python
p1 = data[data['Medal'] == 'Gold'].groupby(by=['Country']).count()['Medal']
p1 = p1.sort_values(ascending = False)
x_1 = list(p1.keys())
y_1 = list(p1.values)
```

```python
fig_top10 , ax_top10 = plt.subplots(figsize = (10,6))
plt.bar(x_1[:10] , y_1[:10] , color = "#005f73"  )
ax_top10.set_facecolor("#EAF4f4")
for spine in ['top' , 'left' , 'right'] : 
    ax_top10.spines[spine].set_visible(False)
ax_top10.tick_params(axis = 'x' , length = 0 , labelrotation = 45)
plt.title("Top 10 Countries with the Most Gold Medals" , fontsize = 12 , color = "#22333b")
plt.ylabel("Number of Gold Medals" , fontsize = 10 , color = "#22333b")
plt.yticks([500*i for i in range(int(max(y_1)/500)+1)])
plt.show()
```

```python
ff2 = data.groupby(by=['Year' , 'Gender']).count()['Medal']
key2 = ff2.keys()
medals_2 = list(ff2.values)
x_2 = []
men_2 = []
women_2 = []
for i in range(len(key2)) : 
    x_2.append(key2[i][0])
    men_2.append(ff2[(key2[i][0],'Men')])
    if i == 0 :
        women_2.append(0)
    else : 
        women_2.append(ff2[(key2[i][0],'Women')])
```

```python
fig_gen , ax_gen = plt.subplots(figsize=(10,6))
ax_gen.set_facecolor("#EAF4f4")
fig_gen.set_facecolor("#EAF4f4")
plt.plot(x_2 , men_2 , color = "#005f73" ,  marker = "o" )
plt.plot(x_2 , women_2 , color = "#94d2bd" , marker = "o")
plt.fill_between(x_2 , women_2 , y2=0 , color = "#94d2bd" , alpha = 0.6)
plt.fill_between(x_2 , men_2 , women_2  , color = "#005f73" , alpha = 0.6)
plt.title("Male and Female Won Medals over the Years" , fontsize = 12 , color = "#22333b" )
plt.ylabel("Number of Won Medals", fontsize = 10 , color = "#22333b")
plt.tick_params( length = 0 , color = "#22333b")
for spine in ['top' , 'left' , 'right'] : 
    ax_gen.spines[spine].set_visible(False)
ax_gen.yaxis.tick_right()
plt.xlim([1896, 2012])
plt.legend(['Men' , 'Women'] , facecolor = "#EAF4f4")
plt.show()
```

```python
dd3 = data[data['Country'] == 'SVK'].groupby(by=['Sport']).count()['Year']
key3 = list(dd3.keys())
sports_1 = {}
for i in range(len(key3)) : 
    sports_1[key3[i]] = dd3[key3[i]] / sum(dd3.values)
```

```python
dd4 = data[data['Country'] == 'SVK'].groupby(by=['Sport' , 'Medal']).count()['Year']
key4 = list(dd4.keys())
medals_4 = []
vals_4 = []
for i in range(len(key4)) : 
    ct  = key4[i][0]
    medals_4.append(key4[i][1])
    vals_4.append(dd4[key4[i]]/sum(dd4.values))
```

```python
outer_colors = ['#d8f3dc', '#b7e4c7', '#95d5b2','#74c69d','#52b788','#40916c']	
```

```python
inner_colors = []
ccs = {'Gold': '#E9C46A', 'Silver': '#E0E1DD', 'Bronze': '#CD7F32'}
for i in medals_4 : 
    inner_colors.append(ccs[i])
```

```python
fig_SVK, ax_SVK = plt.subplots(figsize = (10,8))
plt.pie(sports_1.values() , radius= 1 , colors = outer_colors , labels = key3 
        , wedgeprops=dict(width= 0.3 ,  edgecolor='w'))
plt.pie(vals_4, radius=0.7 , colors = inner_colors , labels = medals_4 , labeldistance = 0.75
       , wedgeprops=dict(width= 0.3 ,  edgecolor='w'))
fig_SVK.set_facecolor("#EAF4f4")
plt.title("Count of Medals in each Sport" , fontsize= 12 , color = "#22333b")
plt.show()
```

```python
import zlib
import zipfile
import joblib

joblib.dump(fig_gen, "fig_gen")
joblib.dump(fig_top10, "fig_top10")
joblib.dump(fig_SVK, "fig_SVK")

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

file_names = ["fig_gen", "fig_top10", "fig_SVK", "Olympic.ipynb"]
compress(file_names)
```

```python
mean_pop = data.mean().values[0]
std_pop = data.std().values[0]
print(mean_pop , std_pop)
```



<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
اکنون ۱۰۰۰ مرتبه نمونه‌گیری به اندازه‌ی ۳۰ عدد از جمعیت بگیرید و به سوالاتی که در ادامه پرسیده می‌شود پاسخ دهید. 
</font>
</p>

```python
#TO-DO
all_samples = []
n_samples = 1000
for i in range(n_samples) : 
    all_samples.append(np.random.choice(data[0].values, size=(30)))
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
میانگین و انحراف از معیار هر ست از نمونه‌ها را محاسبه کرده و در متغیر‌های
<code>mean_sample_set</code>
و
<code>std_sample_set</code>
ذخیره کنید.
</font>
</p>

```python
#TO-DO
mean_sample_set = []
std_sample_set = []
for sample in all_samples : 
    mean_sample_set.append(sample.mean())
    std_sample_set.append(sample.std())
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
 هیستوگرام توزیع نمونه‌‌ها در ۱۰ نمونه‌برداری اول را در قالب ۱۰ تصویر رسم نمایید. 

</font>
</p>

```python
#TO-DO
n_row_1 = 5
n_col_1 = 2
fig1 , ax1 = plt.subplots(figsize = (20,20) , nrows = n_row_1, ncols = n_col_1)
for i in range(n_row_1) : 
    for j in range(n_col_1) : 
        k = i*n_col_1 + j
        ax1[i][j].hist(all_samples[k] , bins= 10)
        ax1[i][j].set_title(f"sample {k+1}")
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
اکنون هیستوگرام توزیع میانگین تمامی ۱۰۰۰ نمونه‌برداری‌ را در یک تصویر رسم نمایید. آیا تصویر به‌دست آمده قابل پیش‌بینی بود؟
</font>
</p>

```python
#TO-DO
fig2 , ax2 = plt.subplots(figsize=(10,10))
all_extend = []
for sample in all_samples : 
    all_extend.extend(sample)
# ax2.hist(all_extend , bins=100)
# ax2_2 = ax2.twinx()
# sns.kdeplot(all_extend)
# plt.ylabel("number of samples")
# plt.xlabel("data")
# plt.title("distribution of data")
ax2.hist(mean_sample_set , bins = 100)
plt.title("distribution of means")
plt.xlabel("means")
plt.ylabel("number of means")
plt.show()
```



<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
تقریبا ۹۵ درصد میانگین نمونه‌ها در چه بازه‌ای قرار می‌گیرد؟ (توجه: این تقریب را بر اساس نرمال فرض کردن توزیع این نمونه‌ها درنظر می‌گیریم.)</font>
</p>

```python
#TO-DO
means_df = pd.Series(mean_sample_set)
lower = means_df.quantile(0.05)
upper = means_df.quantile(0.95)
print(f"lower : {lower} , upper : {upper}")
```


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
بازه‌ی اطمینان ۹۵ درصد در قسمت قبل را با استفاده از z-table محاسبه کنید.
</font>
</p>


```python
#TO-DO
mean_sample_set = sorted(mean_sample_set)
zscores = stats.zscore(mean_sample_set)
inter = (zscores > -1.96) & (zscores < 1.96)
interval = np.array(mean_sample_set)[inter]
upper_bound = interval[-1]
lower_bound = interval[0]
print(f"lower bound : {lower_bound} , upper bound : {upper_bound}")
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای هر ۱۰۰۰ نمونه‌گیری صورت گرفته در بالا(بخش اول)، بازه‌ی اطمینان ۹۵ درصد را حساب کنید. درصد تعداد نمونه‌هایی که میانگین جمعیت در بازه‌ی اطمینان ۹۵ درصد آن‌ها(نمونه‌ها) در بازه‌ی بدست آمده بوده‌اند را گزارش کنید.
</font>
</p>


```python
smpls_out = []
smpls_out_means = 0
def get_95_interval(sample) : 
    score = stats.zscore(sample)
    inter = (score > -1.96) & (score<1.96)
    interval = np.array(sample)[inter]
    lower = interval[0]
    upper = interval[-1]
    mean = interval.mean()
    return [lower , upper , mean]
for i in range(n_samples) : 
    int_95 = get_95_interval(all_samples[i])
    smpls_out.append(int_95[0:2])
    mean = int_95[2]
    if mean >= lower_bound and mean <= upper_bound : 
        smpls_out_means += 1
ress = smpls_out_means / n_samples
print(f"percent of samples that their 95% interval is in the 95% interval of all samples : {ress}")
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
درس جدیدی آماده کرده‌ایم که شامل ۶۰ دقیقه ویدیو است. می‌خواهیم این درس را در دوره‌ی آموزشی دانشجویان بگنجانیم. اما در این درس تغییر استراتژی داده‌ایم و می‌خواهیم ببینیم این تغییر نرخ مشارکت را بیشتر می‌کند یا نه. 
به ۲۰ دانشجو آن را عرضه کرده‌ایم و میانگین نرخ مشارکت آن‌ها ۰.۱۳ بود. 
اگر بخواهیم این درس را برای همه دانشجویان این جمعیت آماده کنیم، برآورد نقطه‌ای و بازه‌ای شما برای نرخ مشارکت دانشجویان بر اساس نمونه‌ای با اندازه‌ی ۲۰ چقدر است؟ پاسخ شما باید برای حداقل ۹۸ درصد جمعیت معتبر باشد.
</font>
</p>

```python
#TO-DO
import math
u_x = 0.13
std_x = 0.15
z_x = 0.5040
n_x = 20
low_x = u_x - (z_x*std_x/math.sqrt(n_x))
up_x = u_x + (z_x*std_x/math.sqrt(n_x))
print(f"lower bound : {low_x} , upper bound : {up_x}")
```

</font>

```python
import zlib
import zipfile

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

file_names = ["class_participation.ipynb"]

compress(file_names)
```
